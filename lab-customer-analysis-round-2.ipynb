```python
#dealing with nan() values
```

```python
file1.isna().sum()
```

```python
#since column months_since_last_claim is numerical.
# We will check the first 50 rows in order to decide our strategy
(file1['months_since_last_claim']).head(50)
```

```python
#since there are no outliers we can replace null values in the numerical data with the mean value
mean_months_since_last_claim = file1['months_since_last_claim'].mean()
mean_months_since_last_claim
```

```python
file1 ['months_since_last_claim'] = file1 ['months_since_last_claim'].fillna(mean_months_since_last_claim)
file1
```

```python
file1.isna().sum()
```

```python
##since there are no outliers we can replace null values in the numerical data with the mean value
file1['number_of_open_complaints'].head(60) 
```

```python
#since it is numerical type median is best soultion to replace nuns if there are outliers since there are no outliers we use here mean
mean_number_of_open_complaints = file1['number_of_open_complaints'].mean()
mean_number_of_open_complaints
```

```python
#replace null values with the mean
file1['number_of_open_complaints'] = file1['number_of_open_complaints'].fillna(mean_number_of_open_complaints)
```

```python
file1.isna().sum()
```

```python
#since there are not a lot nulls we can filter them
file1['vehicle_class'].value_counts
```

```python
file1 = file1[file1['vehicle_class'].isna()==False]
```

```python
file1.isna().sum()
```

```python
file1[file1['vehicle_size'].isna()==True]
```

```python
file1['vehicle_type'].unique()
file1['vehicle_type'].value_counts()

file1['vehicle_type'].value_counts(dropna=False)
```

```python
#since there is only one category of vehicle size we can replace the null values with A
file1['vehicle_type'] = file1['vehicle_type'].fillna('A')
file1['vehicle_type'].isna().sum()
```

```python
file1['vehicle_type'].value_counts()
```

```python
#Finally we have elimated all the Nan values
file1.isna().sum()

```
